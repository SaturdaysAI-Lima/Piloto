```
import pandas as pd

# Read the files
plazavea_electro = pd.read_csv("/content/drive/My Drive/plazavea_electro_v2.csv")
tottus_electro = pd.read_csv("/content/drive/My Drive/tottus_electrodomesticos.csv")

#Elimina columnas sobrantes
def clean_columns(data_set):
  for h in data_set.columns:
    if h == "web-scraper-order":
      data_set.drop(columns="web-scraper-order",axis=1,inplace=True)
    elif h == "web-scraper-start-url":
      data_set.drop(columns="web-scraper-start-url",axis=1,inplace=True)
    elif h == "Categorias":
      data_set.drop(columns="Categorias",axis=1,inplace=True)
    elif h == "Categorias-href":
      data_set.drop(columns="Categorias-href",axis=1,inplace=True)
      return data_set
    
# Agrega el nombre del retail a las columnas
def rename_columns(data_set, name):
  for f in data_set.columns:
     if f.find(name) < 0:
      data_set.rename(columns={f:f + ' ' + name},inplace=True)
  return data_set

clean_columns(plazavea_electro)
rename_columns(plazavea_electro,'Vea')
clean_columns(tottus_electro)
rename_columns(tottus_electro,'Tottus')

new_comp = pd.DataFrame(tottus_electro.head(1))
new_comp.index = range(2,2+1)
new_df = plazavea_electro.join(new_comp)
new_comp.index = range(1,1+1)
new_df.update(new_comp)


print(len(plazavea_electro))
print(len(tottus_electro))
print(len(new_df))
new_df.head(3)
```
