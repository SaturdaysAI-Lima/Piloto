```
#Document Similarity using Euclidean Distance
#https://www.youtube.com/watch?v=Dd16LVt5ct4 (explanation)
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import euclidean_distances
import pandas as pd
```

```
# Create one list of items
corpus = ['Roberto va a la iglesia',
          'Ricardo va a la iglesia',
          'Ha Diana le gusta ir a pescar con sus padres',
          'Ha Silvana le gusta comer ceviche']

# Vectorizing
vectorizer = CountVectorizer()
features = vectorizer.fit_transform(corpus).todense()
print(vectorizer.vocabulary_)
print(features)
```

```
# Loop - comparing one item
# https://en.wikipedia.org/wiki/Euclidean_distance
for f in features:
  print(euclidean_distances(features[0],f))
```
