```
import re  #Regular expression operations
import requests
```

```
# Definir Url de la página que vamos a Scrapear
#url = "https://www.tottus.com.pe/tottus/browse/Chompas/cat1590016"
url = "https://www.tottus.com.pe/tottus/browse/Electrodom%C3%A9sticos/cat880032"
```

```
''' 
GET should not be used for operations that cause side-effects, such as using it for taking actions in web applications. 

POST submits data to be processed (e.g., from an HTML form) to the identified resource. The data is included in the body of the request.
This may result in the creation of a new resource or the updates of existing resources or both.
'''
response = requests.get(url)
print(response.content)
```

```
response_str = response.content.decode("utf-8") # Unicode Transformation Format 8-bytes.
print(response_str)
```

```
# findall() matches all occurrences of a pattern, not just the first one as search() does.
# https://www.w3schools.com/python/python_regex.asp
# Obtener el nombre del producto

product1 = re.findall(r'<div class="title">\n.+\n.+\n.+\n.+</div>', response_str)
product2 = str(product1[0]).split('/')[4]  #split() parte el string en varios strings separados por el valor que se defina



#print(len(product1))
#for i in range (0,2):#len(product1)):
#  print(product1[i])
#print(product2)

# Search() Scan through string looking for the first location where the regular expression pattern produces a match, and return a corresponding match object.
txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt)
if (x):
  print(str(x).split('/s'))
else:
  print("No match")

#product3 = re.search(r'<div class="title">\n.+\n.+\n.+\n.+</div>', response_str)
#print(product3)
```
